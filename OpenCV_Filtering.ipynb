### 직접 커널을 생성하여 필터 적용하기

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('frogimg.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

'''
4x4 커널 
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
1/16 1/16 1/16 1/16
'''
#Basic Blurring 커널을 직접 만들어 보았다.
size = 4
kernel = np.ones((size, size), np.float32) / (size**2)
print(kernel)

dst = cv2.filter2D(image, -1, kernel)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

### Basic Blurring

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('frogimg.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# 4 X 4커널을 자동으로 만들어 준다.
# 커널 사이즈 값이 커질수록 전체에 적용되어
# 블러링 효과가 커진다.
dst = cv2.blur(image, (5, 5))
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

### Gaussian Blur

```
import cv2

image = cv2.imread('frogimg.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# kernel_size: 홀수만 가능
# 커널 사이즈 값이 커질수록 전체에 적용되어
# 블러링 효과가 커진다.
dst = cv2.GaussianBlur(image, (21, 21), 0)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
